```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 500)
```

```python
#load csv file
df = pd.read_csv('file:///Users/DeLaLuna/Downloads/Berlin_crimes.csv')

#add new column and populate with population information
df['City_population'] = np.nan
df.loc[df['Year'] == 2012, 'City_population'] = 3476000
df.loc[df['Year'] == 2013, 'City_population'] = 3488000
df.loc[df['Year'] == 2014, 'City_population'] = 3501000
df.loc[df['Year'] == 2015, 'City_population'] = 3514000
df.loc[df['Year'] == 2016, 'City_population'] = 3526000
df.loc[df['Year'] == 2017, 'City_population'] = 3539000
df.loc[df['Year'] == 2018, 'City_population'] = 3552000
df.loc[df['Year'] == 2019, 'City_population'] = 3557000
df['City_population'] = df['City_population'].astype(int)
```

```python
import pandas as pd


robbery_counts_per_year = df[df['District'] == 'Mitte'].groupby('Year')['Robbery'].sum()

print(robbery_counts_per_year)

```

```python
#display a breakdown of different crime types per year
yearly_crime = df.groupby('Year')[df.columns[4:19]].sum()
yearly_crime
```

```python
#display the yearly overall crimes reported per year
yearly_crime['Total_Crimes'] = yearly_crime.sum(axis=1)
print(yearly_crime['Total_Crimes'])
```

```python
#display the crimes per district

district_crime = df.groupby('District')[df.columns[4:19]].sum()
district_crime
```

```python
#find overall total per district and rank in descending order

district_totals = df.groupby('District')[df.columns[4:19]].sum().sum(axis=1)
district_totals = district_totals.sort_values(ascending=False)
print(district_totals)
```

```python
#input data obtained from external source into a new column called 'city_population'
#Citations for midterm project
#1. <a href='https://www.macrotrends.net/cities/204296/berlin/population'>Source</a>. — Berlin population demographics 2014 - 2019
import numpy as np

conditions = [
    df['Year'] == 2012,
    df['Year'] == 2013,
    df['Year'] == 2014,
    df['Year'] == 2015,
    df['Year'] == 2016,
    df['Year'] == 2017,
    df['Year'] == 2018,
    df['Year'] == 2019
]

values = [3476000, 3488000, 3501000, 3514000, 3526000, 3539000, 3552000, 3557000]

df['City_population'] = np.select(conditions, values, default=np.nan).astype(int)
```

```python
import math
#show the year on year change in population, total crimes and crime per 1000 population

df['City_population'] = df.groupby('Year')['City_population'].transform('first')
yearly_crime['Crimes per 1000'] = (yearly_crime['Total_Crimes'] / yearly_crime['City_population']) * 1000
yearly_crime['Crimes per 1000'] = yearly_crime['Crimes per 1000'].apply(math.ceil)
output_df = yearly_crime[['Year', 'City_population', 'Total_Crimes', 'Crimes per 1000']].drop_duplicates('Year')
print(output_df)

```

```python
# Create new columns for crime types
df['Larceny_type'] = df['Theft'] + df['Car'] + df['From_car'] + df['Bike']
df['Property_damage_type'] = df['Damage'] + df['Graffiti']
df['Violent_crimes_type'] = df['Robbery'] + df['Street_robbery'] + df['Threat'] + df['Agg_assault']
df['Drugs_type'] = df['Drugs']
df['Pyro_type'] = df['Arson'] + df['Fire']
df['Injury_type'] = df['Injury']
df = df.drop('Crime_Type', axis=1)

df
```
